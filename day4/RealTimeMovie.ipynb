```python
from bs4 import BeautifulSoup  
import pandas as pd
```

```python
from urllib.request import urlopen
mpage= urlopen("https://movie.naver.com/movie/sdb/rank/rmovie.nhn")
soup = BeautifulSoup(mpage, "html.parser")
soup
```

```python
ranking=soup.select('#reserveRanking0  a') # "#"이 붙으면 ID를 찾는다; ID가 reserveRanking0인 바디 안에서  a라는 태그를 찾는다
for title in ranking: # a라는 태그 안에서 
    print(title.text) # text를 찾아서 출력한다
```
