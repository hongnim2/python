# NumPy를 좀 더 쉽게 사용하도록 만들어진게 Pasdas

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame # 1차원, 2차원 객체 클래스
```

#Series Test (Class)

```python
arr=np.random.randint(5,20,5) #5~19 까지 임의의 값 5개 발생
print(arr)

s1=Series(arr)
print(s1)

# Labeling;customized indexing
s2=Series(arr, index=list("abcde"))
print(s2)

print(s2['b'])
s2['b']=12
print(s2)

# slicing
print(s2[1:3]) # index일 때는 low포함, high제외
print(s2['b':'d']) # label일 때는 low포함, high포함
```

```python
s2[s2>10] # 각 셀에 조건식 적용하여 표출
```

```python
s3=s2<11 # 각 셀에 조건식 적용하여 참/거짓 표출
print(s3)
```

```python
s3=s2.copy()
s3[s3<11]=0 # 각 셀에 조건식 적용하여 값 쓰기
print(s3)
```

```python
s4=s2*10    # 각각의 데이터에 ""*10" 이 적용된다
print(s4)
```

```python
s2.sort_values(ascending=False)
```

```python
s2.sort_index()
```
