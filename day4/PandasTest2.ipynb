```python
people=pd.read_excel("People.xls",
                     header=2,
                     usecols='B,D,G,J,N', #자치구(B), 인구합계(D), 한국인합계(G), 외국인합계(J), 고령자(N)
                     encoding="utf-8")
people.columns=['구별','인구수','한국인','외국인','고령자']
people.head(2)
```

```python
# 1.Index를 구별로 변경
people.set_index(['구별'], inplace=True)
people.head(6)
```

```python
# 2.인구수가 많은 순서부터 출력
people.sort_values(by='인구수', ascending=False, inplace=True)
people.head(5)
```

```python
# 3. 구별 외국인 비율을 구하세요.
people['외국인비율']=people['외국인']/people['인구수']*100
people.head(3)
```

```python
# 4.인구수를 삭제
del people['인구수']
people.head(3)
```

```python
# 5.합계 줄을 삭제
people.drop('합계',inplace=True)
people.head(4)
```

```python
# 6.외국인 수가 10,000명 이상인 구만 출력
people[people['외국인']>10000]
```
